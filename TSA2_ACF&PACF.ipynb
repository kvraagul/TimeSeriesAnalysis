```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/ice_cream.csv')
```

```python
df.head()
```

```python
df.rename(columns={'IPN31152N':'production'},inplace=True)
```

```python
df['DATE'] = pd.to_datetime(df.DATE)
```

```python
df.set_index('DATE',inplace=True)
```

```python
df.head(3)
```

```python
start_date = pd.to_datetime('2010-01-01')
df = df[start_date:]
```

```python
df.head(3)
```

```python
df.tail()
```

```python
plt.plot(df)
plt.rcParams['figure.figsize']=(15,5)
plt.title('Ice Cream Production from 2010-2020',fontsize=15)
```

# ACF and PACF

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

## ACF

```python
acf_plot = plot_acf(df.production)
```

```python
acf_plot1 = plot_acf(df.production, lags=100)
```

## PACF

```python
pacf_plot = plot_pacf(df.production)
```
